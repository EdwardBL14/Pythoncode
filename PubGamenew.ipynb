```python
def Bethtotal(array,pairpicked,first_emptyadded,second_emptyadded):
    n = int(len(array))
    i = pairpicked
    i2 = (i+1)%n #to get i and i2 as two adajcent glasses in the cirlular arrangement
    array = np.delete(array, [[i, i2]])
    array = np.insert(array, first_emptyadded, 0)
    array = np.insert(array, second_emptyadded, 0)
    return array
```

```python
def Alistrat(array,turn): #Ali's strategy for n = 8
    
    n = len(array)
    divider = 5-turn
    amount_added = (1/2)/divider
    non_zero_indices = np.nonzero(array)
    
    if len(non_zero_indices) != divider:
        placement_differences = np.diff(non_zero_indices)
        diff = placement_differences[-1] - placement_differences[0] - n +2
        placement_differences = np.append(placement_differences, diff)
        indices_of_ones = np.where(placement_differences == 1)[0]
        non_zero_indices = np.delete(non_zero_indices, indices_of_ones)
        
    for k in range(len(non_zero_indices)):
        location = non_zero_indices[k]
        #value = array[location]
        array[location] = array[location] + amount_added
    
    max_value = np.max(array)
    if turn == 4 and max_value <= 1:
        print('Strategy Failed')
        print(array)
    return array
```
