```python
import numpy as np
import random as random
import matplotlib.pyplot as plt
import seaborn as sns

array = [1,0.5]
array = np.array(array)

for k in range(23):
    x1 = array[k]
    x2 = array[k+1]
    x3 = 0.5*x1 + 0.5*x2
    array = np.append(array,x3)
    
x = np.linspace(1, 25,25)
twothirds = np.full((25,), 2/3)
    
plt.plot(x,array,label = 'Probability of reaching each timestep')
plt.plot(x,twothirds,label = '2/3',linestyle='--',color='red')
plt.xlabel("step n", fontsize=10)
plt.ylabel("Probability", fontsize=10)
plt.legend(loc='upper right', fontsize='large')
plt.show()
```

```python
vals = [0]
vals = np.array(vals)

for k in range(10000):
    array = [1]
    array = np.array(array)
    for k in range(25):
        x = random.randint(1, 2)
        u = int(array[k])
        new = u + x
        array = np.append(array,new)
    vals = np.append(vals,array)

s_2 = np.count_nonzero(vals == 2)
print(s_2/10000)
s_25 = np.count_nonzero(vals == 25)
print(s_25/10000)
```

```python
array2 = [1,1/3,4/9]
array2 = np.array(array2)

for k in range(22):
    x1 = array2[k]
    x2 = array2[k+1]
    x3 = array2[k+2]
    x4 = (x1+x2+x3)/3
    array2 = np.append(array2,x4)
    
x = np.linspace(1, 25,25)
twothirds = np.full((25,), 0.5)
    
plt.plot(x,array2,label = 'Probability of reaching each timestep')
plt.plot(x,twothirds,label = '1/2',linestyle='--',color='red')
plt.xlabel("step n", fontsize=10)
plt.ylabel("Probability", fontsize=10)
plt.legend(loc='upper right', fontsize='large')
plt.show()
```
